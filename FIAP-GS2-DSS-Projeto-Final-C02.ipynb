### Pegada de Carbono

A pegada de carbono de um país refere-se à quantidade total de emissões de gases de efeito estufa (GEE) associadas à sua economia, população e atividades. Essa medida inclui as emissões diretas e indiretas de dióxido de carbono (CO2), metano (CH4), óxido nitroso (N2O) e outros gases, resultantes da queima de combustíveis fósseis, produção industrial, agricultura, desmatamento, transporte, consumo de energia e outros setores. A pegada de carbono é calculada considerando as emissões territoriais, ou seja, as emitidas dentro das fronteiras do país, e as emissões de consumo, que incluem as emissões associadas à produção de bens e serviços consumidos pelo país, independentemente de onde foram produzidos. Isso significa que a pegada de carbono de um país pode ser influenciada por fatores como o tipo de energia utilizada, a eficiência energética, a produção e consumo de alimentos, a indústria manufatureira, o transporte e a logística, além da política climática e regulatória do governo. A pegada de carbono é uma métrica importante para avaliar o impacto ambiental de um país e monitorar seu progresso em direção à redução de emissões e ao cumprimento dos objetivos do Acordo de Paris, que visa limitar o aquecimento global a menos de 2°C até 2050. Além disso, a pegada de carbono pode ser utilizada para orientar políticas públicas e estratégias de mitigação e adaptação às mudanças climáticas, bem como para promover práticas sustentáveis e responsáveis entre empresas, governos e cidadãos.

Algumas estatísticas importantes:

- Em 2020, os 10 maiores emissores de CO2 foram: China (25%), EUA (14%), Índia (7%), Rússia (5%), Japão (4%), Alemanha (3%), Irã (3%), Coreia do Sul (2%), Canadá (2%) e Indonésia (2%).
- A pegada de carbono per capita varia significativamente entre os países, com os EUA apresentando uma das maiores pegadas per capita (16,4 toneladas de CO2 equivalente por pessoa em 2020).
- A redução da pegada de carbono requer esforços coordenados em setores-chave, como energia renovável, eficiência energética, eletrificação do transporte, agricultura sustentável e gestão de resíduos.

### Separando os países em conjuntos úteis

```python
oecd = {'Australia', 'Austria', 'Belgium', 'Canada', 'Chile', 'Czechia', 'Denmark', 
        'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 
        'Ireland', 'Israel', 'Italy', 'Japan', 'Luxembourg', 'Mexico', 'Netherlands', 
        'New Zealand', 'Norway', 'Poland', 'Portugal', 'Slovakia', 'Slovenia', 
        'South Korea', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom', 
        'United States'}

north_america = {'Canada', 'Mexico', 'United States'}

south_america = {'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 
               'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela'}

central_america = {'Belize', 'Costa Rica', 'El Salvador', 'Guatemala', 'Honduras', 
                  'Nicaragua', 'Panama'}

americas = {'North America', 'South America', 'Central America'}
```

```python
def aggregate_group(df, country_list, group_name):
    # Filtrar os países do grupo
    group_data = df[df['country'].isin(country_list)]
    
    # Agrupar por ano e calcular as somas/médias
    aggregated = group_data.groupby('year').agg({
        'co2': 'sum',  # Soma para emissões totais
        'co2_per_capita': 'mean',  # Média para emissões per capita
        'cumulative_co2': 'sum',  # Soma para emissões cumulativas
        'population': 'sum',  # Soma para população
        'gdp': 'sum',  # Soma para PIB
        'cumulative_co2': 'sum',
        'cumulative_co2_including_luc': 'sum',
        'co2_per_gdp': 'mean',
        'energy_per_capita': 'mean',
        'coal_co2': 'sum',
        'gas_co2': 'sum',
        'oil_co2': 'sum',
        'cement_co2': 'sum'
        # Adicione outras colunas relevantes aqui
    }).reset_index()
    
    # Adicionar o nome do grupo como "país"
    aggregated['country'] = group_name
    
    return aggregated
```

```python
oecd_data = aggregate_group(df, oecd, 'OECD')
north_america_data = aggregate_group(df, north_america, 'North America')
south_america_data = aggregate_group(df, south_america, 'South America')
central_america_data = aggregate_group(df, central_america, 'Central America')
```

```python
df_grouped = pd.concat([df, oecd_data, north_america_data, south_america_data, central_america_data], ignore_index=True)
```

```python
americas_data = aggregate_group(df_grouped, americas, 'Americas')
```

```python
df_grouped = pd.concat([df_grouped, americas_data], ignore_index=True)
```

3. Clustering Model: Segmenting Countries by Emission Patterns
Model Type: K-Means Clustering
Question: How can countries, including Brazil, be grouped based on their emission profiles and energy consumption patterns?
Features: co2_per_capita, co2_per_gdp, energy_per_capita, share_global_co2, cement_co2, coal_co2, gas_co2, oil_co2, land_use_change_co2
Objective: Identify clusters of countries with similar emission behaviors to tailor regional or global climate policies.

```python
# Agregando os dados por país e calculando a média de cada variável de emissão e consumo de energia
df_aggregated = df_grouped.groupby('country')[features].mean().reset_index()

# Verificando as primeiras linhas após a agregação
df_aggregated.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados (substitua por sua própria fonte de dados)
# brazil_model_df = pd.read_csv('dados.csv')

# Variáveis de interesse
features = [
    'co2_per_capita', 'co2_per_gdp', 'energy_per_capita', 'share_global_co2',
    'cement_co2', 'coal_co2', 'gas_co2', 'oil_co2', 'land_use_change_co2'
]

# Selecionando apenas as colunas de interesse
df = df_aggregated[features]

# Preenchendo valores ausentes com a média da coluna
df_filled = df.fillna(df.mean())

# Normalizando os dados
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_filled)

# Determinando o número ideal de clusters utilizando o método do cotovelo
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plotando o gráfico do método do cotovelo
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Método do Cotovelo")
plt.xlabel("Número de Clusters")
plt.ylabel("Inércia")
plt.show()

# Definindo o número de clusters com base no cotovelo (suponhamos que 4 clusters)
n_clusters = 4

# Aplicando o K-Means com o número ideal de clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_filled['Cluster'] = kmeans.fit_predict(df_scaled)

# Visualizando os clusters no gráfico
plt.figure(figsize=(10, 8))
sns.scatterplot(x=df_filled['co2_per_capita'], y=df_filled['energy_per_capita'], hue=df_filled['Cluster'], palette='viridis', s=100, edgecolor='black')
plt.title("Clusters de Países por Padrões de Emissões e Consumo de Energia")
plt.xlabel('CO2 per Capita')
plt.ylabel('Energy per Capita')
plt.legend(title='Cluster')
plt.show()

# Avaliando o modelo com o Silhouette Score
silhouette_avg = silhouette_score(df_scaled, df_filled['Cluster'])
print(f"Silhouette Score: {silhouette_avg:.2f}")

# Exibindo os países em cada cluster (supondo que 'country' seja a coluna com os nomes dos países)
df_filled['Country'] = df_aggregated['country']
for cluster_num in range(n_clusters):
    print(f"\nCluster {cluster_num}:")
    print(df_filled[df_filled['Cluster'] == cluster_num]['Country'].values)
```

Target: co2_including_luc
Objective: Capture temporal patterns to detect seasonal variations and anomalies in emissions, allowing better planning for emission control.

```python
# Usando 'year' para criar sequências temporais
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Selecionando as colunas relevantes
df_time_series = brazil_model_df[['year', 'co2_including_luc', 'primary_energy_consumption', 'gdp', 'population']]

# Ordenando os dados por ano
df_time_series = df_time_series.sort_values('year')

# Normalizando as variáveis para LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_time_series[['co2_including_luc', 'primary_energy_consumption', 'gdp', 'population']])

# Função para criar as sequências de dados
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i + seq_length]
        y = data[i + seq_length, 0]  # 'co2_including_luc' é o alvo
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

# Definindo o tamanho da sequência (e.g., 4 trimestres = 1 ano)
seq_length = 4

# Criando as sequências
X, y = create_sequences(scaled_data, seq_length)

# Dividindo os dados em treino e teste (80% treino, 20% teste)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape para ser compatível com LSTM (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2])

# Verificando as dimensões
print(X_train.shape, y_train.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Construindo o modelo LSTM
model = Sequential()

# Camada LSTM
model.add(LSTM(50, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))  # Dropout para evitar overfitting

# Camada densa de saída
model.add(Dense(1))

# Compilando o modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Resumo do modelo
model.summary()
```
