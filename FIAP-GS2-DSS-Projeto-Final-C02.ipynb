### Pegada de Carbono

A pegada de carbono de um paÃ­s refere-se Ã  quantidade total de emissÃµes de gases de efeito estufa (GEE) associadas Ã  sua economia, populaÃ§Ã£o e atividades. Essa medida inclui as emissÃµes diretas e indiretas de diÃ³xido de carbono (CO2), metano (CH4), Ã³xido nitroso (N2O) e outros gases, resultantes da queima de combustÃ­veis fÃ³sseis, produÃ§Ã£o industrial, agricultura, desmatamento, transporte, consumo de energia e outros setores. A pegada de carbono Ã© calculada considerando as emissÃµes territoriais, ou seja, as emitidas dentro das fronteiras do paÃ­s, e as emissÃµes de consumo, que incluem as emissÃµes associadas Ã  produÃ§Ã£o de bens e serviÃ§os consumidos pelo paÃ­s, independentemente de onde foram produzidos. Isso significa que a pegada de carbono de um paÃ­s pode ser influenciada por fatores como o tipo de energia utilizada, a eficiÃªncia energÃ©tica, a produÃ§Ã£o e consumo de alimentos, a indÃºstria manufatureira, o transporte e a logÃ­stica, alÃ©m da polÃ­tica climÃ¡tica e regulatÃ³ria do governo. A pegada de carbono Ã© uma mÃ©trica importante para avaliar o impacto ambiental de um paÃ­s e monitorar seu progresso em direÃ§Ã£o Ã  reduÃ§Ã£o de emissÃµes e ao cumprimento dos objetivos do Acordo de Paris, que visa limitar o aquecimento global a menos de 2Â°C atÃ© 2050. AlÃ©m disso, a pegada de carbono pode ser utilizada para orientar polÃ­ticas pÃºblicas e estratÃ©gias de mitigaÃ§Ã£o e adaptaÃ§Ã£o Ã s mudanÃ§as climÃ¡ticas, bem como para promover prÃ¡ticas sustentÃ¡veis e responsÃ¡veis entre empresas, governos e cidadÃ£os.

- Em 2020, os 10 maiores emissores de CO2 foram: China (25%), EUA (14%), Ãndia (7%), RÃºssia (5%), JapÃ£o (4%), Alemanha (3%), IrÃ£ (3%), Coreia do Sul (2%), CanadÃ¡ (2%) e IndonÃ©sia (2%).
- A pegada de carbono per capita varia significativamente entre os paÃ­ses, com os EUA apresentando uma das maiores pegadas per capita (16,4 toneladas de CO2 equivalente por pessoa em 2020).
- A reduÃ§Ã£o da pegada de carbono requer esforÃ§os coordenados em setores-chave, como energia renovÃ¡vel, eficiÃªncia energÃ©tica, eletrificaÃ§Ã£o do transporte, agricultura sustentÃ¡vel e gestÃ£o de resÃ­duos.

### Separando os paÃ­ses em conjuntos Ãºteis

```python
oecd = {'Australia', 'Austria', 'Belgium', 'Canada', 'Chile', 'Czechia', 'Denmark', 
        'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 
        'Ireland', 'Israel', 'Italy', 'Japan', 'Luxembourg', 'Mexico', 'Netherlands', 
        'New Zealand', 'Norway', 'Poland', 'Portugal', 'Slovakia', 'Slovenia', 
        'South Korea', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom', 
        'United States'}

north_america = {'Canada', 'Mexico', 'United States'}

south_america = {'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 
               'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela'}

central_america = {'Belize', 'Costa Rica', 'El Salvador', 'Guatemala', 'Honduras', 
                  'Nicaragua', 'Panama'}

americas = {'North America', 'South America', 'Central America'}
```

```python
def aggregate_group(df, country_list, group_name):
    # Filtrar os paÃ­ses do grupo
    group_data = df[df['country'].isin(country_list)]
    
    # Agrupar por ano e calcular as somas/mÃ©dias
    aggregated = group_data.groupby('year').agg({
        'co2': 'sum',  # Soma para emissÃµes totais
        'co2_per_capita': 'mean',  # MÃ©dia para emissÃµes per capita
        'cumulative_co2': 'sum',  # Soma para emissÃµes cumulativas
        'population': 'sum',  # Soma para populaÃ§Ã£o
        'gdp': 'sum',  # Soma para PIB
        'cumulative_co2': 'sum',
        'cumulative_co2_including_luc': 'sum',
        'co2_per_gdp': 'mean',
        'energy_per_capita': 'mean',
        'coal_co2': 'sum',
        'gas_co2': 'sum',
        'oil_co2': 'sum',
        'cement_co2': 'sum'
        # Adicione outras colunas relevantes aqui
    }).reset_index()
    
    # Adicionar o nome do grupo como "paÃ­s"
    aggregated['country'] = group_name
    
    return aggregated
```

```python
oecd_data = aggregate_group(df, oecd, 'OECD')
north_america_data = aggregate_group(df, north_america, 'North America')
south_america_data = aggregate_group(df, south_america, 'South America')
central_america_data = aggregate_group(df, central_america, 'Central America')
```

```python
df_grouped = pd.concat([df, oecd_data, north_america_data, south_america_data, central_america_data], ignore_index=True)
```

```python
americas_data = aggregate_group(df_grouped, americas, 'Americas')
```

```python
df_grouped = pd.concat([df_grouped, americas_data], ignore_index=True)
```

3. Clustering Model: Segmenting Countries by Emission Patterns
Model Type: K-Means Clustering
Question: How can countries, including Brazil, be grouped based on their emission profiles and energy consumption patterns?

```python
# Agregando os dados por paÃ­s e calculando a mÃ©dia de cada variÃ¡vel de emissÃ£o e consumo de energia
df_aggregated = df_grouped.groupby('country')[features].mean().reset_index()

# Verificando as primeiras linhas apÃ³s a agregaÃ§Ã£o
df_aggregated.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados (substitua por sua prÃ³pria fonte de dados)
# brazil_model_df = pd.read_csv('dados.csv')

# VariÃ¡veis de interesse
features = [
    'co2_per_capita', 'co2_per_gdp', 'energy_per_capita', 'share_global_co2',
    'cement_co2', 'coal_co2', 'gas_co2', 'oil_co2', 'land_use_change_co2'
]

# Selecionando apenas as colunas de interesse
df = df_aggregated[features]

# Preenchendo valores ausentes com a mÃ©dia da coluna
df_filled = df.fillna(df.mean())

# Normalizando os dados
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_filled)

# Determinando o nÃºmero ideal de clusters utilizando o mÃ©todo do cotovelo
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plotando o grÃ¡fico do mÃ©todo do cotovelo
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title("MÃ©todo do Cotovelo")
plt.xlabel("NÃºmero de Clusters")
plt.ylabel("InÃ©rcia")
plt.show()

# Definindo o nÃºmero de clusters com base no cotovelo (suponhamos que 4 clusters)
n_clusters = 4

# Aplicando o K-Means com o nÃºmero ideal de clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_filled['Cluster'] = kmeans.fit_predict(df_scaled)

# Visualizando os clusters no grÃ¡fico
plt.figure(figsize=(10, 8))
sns.scatterplot(x=df_filled['co2_per_capita'], y=df_filled['energy_per_capita'], hue=df_filled['Cluster'], palette='viridis', s=100, edgecolor='black')
plt.title("Clusters de PaÃ­ses por PadrÃµes de EmissÃµes e Consumo de Energia")
plt.xlabel('CO2 per Capita')
plt.ylabel('Energy per Capita')
plt.legend(title='Cluster')
plt.show()

# Avaliando o modelo com o Silhouette Score
silhouette_avg = silhouette_score(df_scaled, df_filled['Cluster'])
print(f"Silhouette Score: {silhouette_avg:.2f}")

# Exibindo os paÃ­ses em cada cluster (supondo que 'country' seja a coluna com os nomes dos paÃ­ses)
df_filled['Country'] = df_aggregated['country']
for cluster_num in range(n_clusters):
    print(f"\nCluster {cluster_num}:")
    print(df_filled[df_filled['Cluster'] == cluster_num]['Country'].values)
```

```python
# Usando 'year' para criar sequÃªncias temporais
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Selecionando as colunas relevantes
df_time_series = brazil_model_df[['year', 'co2_including_luc', 'primary_energy_consumption', 'gdp', 'population']]

# Ordenando os dados por ano
df_time_series = df_time_series.sort_values('year')

# Normalizando as variÃ¡veis para LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_time_series[['co2_including_luc', 'primary_energy_consumption', 'gdp', 'population']])

# FunÃ§Ã£o para criar as sequÃªncias de dados
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i + seq_length]
        y = data[i + seq_length, 0]  # 'co2_including_luc' Ã© o alvo
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

# Definindo o tamanho da sequÃªncia (e.g., 4 trimestres = 1 ano)
seq_length = 4

# Criando as sequÃªncias
X, y = create_sequences(scaled_data, seq_length)

# Dividindo os dados em treino e teste (80% treino, 20% teste)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape para ser compatÃ­vel com LSTM (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2])

# Verificando as dimensÃµes
print(X_train.shape, y_train.shape)
```
